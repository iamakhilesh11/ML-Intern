```python
df['Day_Perc_Change']=df['Close Price'].pct_change() 
```

```python
df
```

```python
df["Day_Perc_Change"] = df["Day_Perc_Change"].fillna(0)
df
```

```python
def trend(Day_Perc_Change):       
        if -0.5 <= df['Day_Perc_Change'] <=  0.5: return 'Slight'
        if  0.5 <= df['Day_Perc_Change'] <=  1:   return 'Slight positive'
        if -0.5 >= df['Day_Perc_Change'] >= -1:   return 'Slight negative'
        if  1   <= df['Day_Perc_Change'] <=  3:   return 'Positive'
        if -1   >= df['Day_Perc_Change'] >= -3:   return 'Negative'
        if  3   <= df['Day_Perc_Change'] <=  7:   return 'Among top gainers'
        if -3   >= df['Day_Perc_Change'] >= -7:   return 'Among top losers'
        if         df['Day_Perc_Change'] >   7:   return 'Bull run'
        if         df['Day_Perc_Change'] <  -7:   return 'Bear drop'
```

```python
df['Trend'] = trend
```
